```python
# 2️⃣ Import and Load BART Model
from transformers import BartTokenizer, BartForConditionalGeneration

# Load pretrained BART model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
```

```python
#3️⃣ Define a Summarization Function
def bart_summarize(text, min_length=30, max_length=200):
    """
    Generate a summary using the BART model.
    """
    if not text.strip():
        return "⚠️ Please enter some text."

    # Tokenize the input
    inputs = tokenizer([text], max_length=1024, truncation=True, return_tensors="pt")

    # Generate summary
    summary_ids = model.generate(
        inputs["input_ids"],
        num_beams=4,
        max_length=max_length,
        min_length=min_length,
        length_penalty=2.0,
        early_stopping=True
    )

    # Decode output
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
# 4️⃣ Try It Out

# You can input any long paragraph and get a summary:
text = """
Artificial intelligence (AI) is transforming industries across the world by enabling
machines to perform tasks that normally require human intelligence. From healthcare
and finance to education and transportation, AI technologies like machine learning,
natural language processing, and robotics are revolutionizing the way we live and work.
However, the rapid advancement of AI also raises important ethical questions regarding
privacy, bias, and employment displacement.
"""

print("📜 Original text length:", len(text.split()), "words\n")
print("🔹 Summary:\n", bart_summarize(text))
```

```python
#5️⃣ Create a Simple Input Box in Colab

# If you’d like a quick UI inside Colab:
from IPython.display import display
import ipywidgets as widgets

text_box = widgets.Textarea(
    value='Paste your text here...',
    placeholder='Enter text to summarize...',
    description='Text:',
    layout=widgets.Layout(width='100%', height='200px')
)

button = widgets.Button(description='Summarize', button_style='success')
output = widgets.Output()

def on_button_click(b):
    with output:
        output.clear_output()
        summary = bart_summarize(text_box.value)
        print("🔹 Summary:\n")
        print(summary)

button.on_click(on_button_click)
display(text_box, button, output)
```
